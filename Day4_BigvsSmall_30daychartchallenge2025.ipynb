```python

#Day 4 of #30DayChartChallenge, Big or Small

import matplotlib.pyplot as plt
import pandas as pd

# Defense budgets in billions USD
data = {
    'Country': ['United States', 'Canada', 'Greenland (Denmark)', 'Mexico'],
    'Budget (Billion USD)': [850, 34.6, 2.05, 9.1]  # Mexico's budget estimate
}

# Create DataFrame
df = pd.DataFrame(data)

# Plot
fig, ax = plt.subplots(figsize=(8, 5))
ax.bar(df['Country'], df['Budget (Billion USD)'], color=['blue', 'red', 'green', 'orange'])

ax.set_ylabel("Defense Budget (Billion USD)")
ax.set_title("Defense Budgets of Selected Countries (2025)")

# Add values on top of bars
for i, value in enumerate(df['Budget (Billion USD)']):
    ax.text(i, value + 2, f"{value}", ha='center', va='bottom', fontsize=10)

plt.show()
```
