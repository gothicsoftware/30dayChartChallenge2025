```
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

def plot_stripes(df, x_col, y_col, title="Stripes Plot"):
    """
    Plots a stripes chart for the given DataFrame.

    Args:
        df (pd.DataFrame): The input DataFrame.
        x_col (str): The column to be used for the x-axis (categorical).
        y_col (str): The column to be used for the y-axis (numerical).
        title (str): The title of the plot.
    """

    plt.figure(figsize=(12, 6))
    sns.stripplot(x=x_col, y=y_col, data=df, jitter=True, alpha=0.7)
    plt.title(title)
    plt.xticks(rotation=45, ha="right")  # Rotate x-axis labels for readability
    plt.tight_layout()
    plt.show()

from google.colab import drive
drive.mount('/content/gdrive')

df = pd.read_csv("/content/gdrive/My Drive/Spotify_2024_Global_Streaming_Data.csv")
# Quick preview
print(df.head())
print(df.info())
```

```
# Stripes plot for Genre vs. Monthly Listeners
plot_stripes(df, "Genre", "Monthly Listeners (Millions)", "Monthly Listeners by Genre")


```

```
# Stripes plot for Platform Type vs. Total Streams
plot_stripes(df, "Platform Type", "Total Streams (Millions)", "Total Streams by Platform")
```

```
# Stripes plot for Country vs. Skip Rate
plot_stripes(df, "Country", "Skip Rate (%)", "Skip Rate by Country")
```
