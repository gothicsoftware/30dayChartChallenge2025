```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Create the dataset
data = {
    "Car Type": [
        "Formula 1 (V10)",
        "Formula 1 (Hybrid V6)",
        "NASCAR",
        "IndyCar",
        "Formula E",
        "Le Mans Prototype (LMP1)",
        "GT3 Sports Car",
        "MotoGP Bike"
    ],
    "Event": [
        "Grand Prix (2000s)",
        "Grand Prix (Current)",
        "Daytona 500",
        "Indianapolis 500",
        "ePrix",
        "24 Hours of Le Mans",
        "IMSA SportsCar",
        "MotoGP Race"
    ],
    "Noise Level (dB)": [140, 120, 130, 128, 80, 115, 120, 115]
}

df = pd.DataFrame(data)

# Set style
sns.set_theme(style="whitegrid")

# Create the bar plot
plt.figure(figsize=(12, 6))
barplot = sns.barplot(
    x="Noise Level (dB)",
    y="Car Type",
    data=df,
    palette="rocket"
)

# Add titles and labels
plt.title("Noise Levels of Different Race Vehicles", fontsize=16, weight='bold')
plt.xlabel("Noise Level (dB)", fontsize=14)
plt.ylabel("Race Car Type", fontsize=14)

# Add dB labels at the end of each bar
for index, value in enumerate(df["Noise Level (dB)"]):
    plt.text(value + 1, index, f"{value} dB", va='center')

plt.xlim(70, 150)
plt.tight_layout()
plt.show()
```
