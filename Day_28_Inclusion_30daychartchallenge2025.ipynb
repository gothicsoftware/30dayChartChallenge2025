Data Source: International Telecommunication Union (ITU), 2024 Global Internet Access Report

https://www.itu.int/itu-d/reports/statistics/2024/11/10/ff24-internet-use-in-urban-and-rural-areas/




```python
import matplotlib.pyplot as plt
import numpy as np


# Data
regions = ['Africa', 'Asia-Pacific', 'Europe', 'Americas', 'Arab States', 'World Avg']
urban_access = [57, 77, 95, 88, 79, 83]
rural_access = [23, 44, 91, 66, 55, 48]

x = np.arange(len(regions))  # label locations
width = 0.35  # bar width

# Create plot
fig, ax = plt.subplots(figsize=(8, 6))
bars1 = ax.bar(x - width/2, urban_access, width, label='Urban', color='#1f77b4')
bars2 = ax.bar(x + width/2, rural_access, width, label='Rural', color='#ff7f0e')

# Add labels, title, and grid
ax.set_ylabel('Internet Access (%)')
ax.set_title('Internet Access by Region and Area Type (2024)')
ax.set_xticks(x)
ax.set_xticklabels(regions)
ax.legend()
ax.grid(axis='y', linestyle='--', alpha=0.7)

# Add text labels on bars
for bars in [bars1, bars2]:
    for bar in bars:
        height = bar.get_height()
        ax.annotate(f'{height}%',
                    xy=(bar.get_x() + bar.get_width() / 2, height),
                    xytext=(0, 3),  # 3 points vertical offset
                    textcoords="offset points",
                    ha='center', va='bottom')

#plt.figtext(0.99, 0.01, 'Source: ITU 2024', horizontalalignment='right', fontsize=8)

# Add a stylish source note

fig.text(0.99, 0.01, 'Source: International Telecommunication Union (ITU), 2024',
         ha='right', va='bottom', fontsize=8, color='gray', alpha=0.7)


plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

# Data
categories = ['Urban Areas', 'Rural Areas']
access_percentages = [83, 48]  # Global internet usage percentages

# Create bar chart
plt.figure(figsize=(8, 6))
bars = plt.bar(categories, access_percentages, color=['#1f77b4', '#ff7f0e'])

# Add data labels
for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2.0, height + 1, f'{height}%', ha='center', va='bottom')

# Chart title and labels
plt.title('Global Internet Access by Area (2024)', fontsize=14)
plt.ylabel('Percentage of Population Using the Internet')
plt.ylim(0, 100)
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Display the chart
plt.tight_layout()
plt.show()
```
