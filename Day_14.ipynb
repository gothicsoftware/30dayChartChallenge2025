```python
import networkx as nx
import matplotlib.pyplot as plt

# Create a directed graph
G = nx.DiGraph()

# === EXAMPLE STRUCTURE ===
# Replace these with your own family names and relationships
# Format: G.add_edge('Parent', 'Child')

# Grandparents generation
G.add_edge('Harry A', 'Paul A')
G.add_edge('Hermione A', 'Paul A')

G.add_edge('Gustavo A', 'Tricia')
G.add_edge('Marisela A', 'Tricia')

# You and Tricia
G.add_edge('Tricia', 'You')

G.add_edge('Jenna A', 'Tricia')
G.add_edge('Kevin A', 'Tricia')

G.add_edge('Glenn S', 'You')

G.add_edge('Gladys S', 'Glenn S')

# Parents generation
G.add_edge('Paul A', 'You')
G.add_edge('Mary H', 'You')



# Your generation
G.add_edge('You', 'Cameron S')
G.add_edge('You', 'Evan S')

# === LAYOUT ===
pos = nx.spring_layout(G, k=1.5, seed=42)  # or use graphviz_layout if you want a top-down tree

# Draw nodes and edges
plt.figure(figsize=(10, 8))
nx.draw(G, pos, with_labels=True, arrows=True, node_color='lightblue',
        node_size=2000, font_size=10, font_weight='bold', edge_color='gray')

plt.title("Smith Family Tree", fontsize=14)
plt.axis('off')
plt.tight_layout()
plt.show()
```

```python
import networkx as nx
import matplotlib.pyplot as plt

# Create a directed graph
G = nx.DiGraph()

# === EXAMPLE STRUCTURE ===
# Format: G.add_edge('Parent', 'Child')

# You and Tricia
G.add_edge('Tricia', 'Paul')
G.add_edge('Paul', 'Tricia')

G.add_edge('Tricia','Jenna A')
G.add_edge('Tricia','Kevin A')

G.add_edge('Gustavo A', 'Tricia')
G.add_edge('Marisela A', 'Tricia')

G.add_edge('Paul', 'Cameron S')
G.add_edge('Paul', 'Evan S')

G.add_edge('Glenn S', 'Paul')
G.add_edge('Mary M', 'Paul')


# === LAYOUT ===
pos = nx.spring_layout(G, k=1.5, seed=42)  # or use graphviz_layout if you want a top-down tree

# Draw nodes and edges
plt.figure(figsize=(10, 8))
nx.draw(G, pos, with_labels=True, arrows=True, node_color='lightblue',
        node_size=2000, font_size=10, font_weight='bold', edge_color='gray')

plt.title("Paul's Family Tree", fontsize=14)
plt.axis('off')
plt.tight_layout()
plt.show()
```
