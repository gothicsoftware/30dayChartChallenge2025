```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Simulated data
urban_data = pd.DataFrame({
    "Country": ["Singapore", "Qatar", "Kuwait", "Belgium", "Japan", "South Korea", "USA", "Brazil", "Mexico", "UK"],
    "Urban_Population_%": [100, 99, 98, 98, 91, 90, 83, 87, 81, 84]
})

plt.figure(figsize=(10, 6))
sns.barplot(x="Urban_Population_%", y="Country", data=urban_data, palette="magma")
plt.title("Top 10 Most Urbanized Countries")
plt.xlabel("Urban Population (%)")
plt.ylabel("")
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import plotly.express as px

# Urbanization data (sample values, feel free to update with real stats)
data = {
    "Country": ["Singapore", "Qatar", "Kuwait", "Belgium", "Japan", "South Korea", "USA", "Brazil", "Mexico", "UK"],
    "Urban_Population_Percent": [100, 99, 98, 98, 91, 90, 83, 87, 81, 84],
    "ISO_Code": ["SGP", "QAT", "KWT", "BEL", "JPN", "KOR", "USA", "BRA", "MEX", "GBR"]
}

df = pd.DataFrame(data)

# Create the choropleth map
fig = px.choropleth(
    df,
    locations="ISO_Code",
    color="Urban_Population_Percent",
    hover_name="Country",
    color_continuous_scale="OrRd",
    range_color=(80, 100),
    title="Top 10 Most Urbanized Countries (by % Urban Population)"
)

fig.update_layout(
    geo=dict(showframe=False, showcoastlines=False),
    coloraxis_colorbar=dict(title="% Urban")
)

fig.show()
```
