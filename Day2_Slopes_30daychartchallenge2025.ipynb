```python
# Load dataset (Assumes CSV format with 'Country', 'Year', 'CO2_Emissions')
df = pd.read_csv("/content/gdrive/My Drive/Total Emissions Per Country (2000-2020).csv")
```

```python


# Filter for CO2 emissions
df_filtered = df[df["Element"] == "Emissions (CO2)"]

# Select relevant columns: 'Area', '2000', and '2020', and drop any rows with missing data
df_pivot = df_filtered[["Area", "2000", "2020"]].dropna()

# Convert emissions data to numeric values
df_pivot["2000"] = pd.to_numeric(df_pivot["2000"], errors='coerce')
df_pivot["2020"] = pd.to_numeric(df_pivot["2020"], errors='coerce')

# Remove non-country entries (e.g., "World", continents, etc.)
invalid_names = ["World", "European Union", "Asia", "Africa", "North America", "South America"]
df_pivot = df_pivot[~df_pivot["Area"].isin(invalid_names)]

# Sort by 2020 emissions and select the top 10 **unique** countries
df_pivot = df_pivot.sort_values(by="2020", ascending=False).drop_duplicates(subset=["Area"]).head(10)

# Plot the slope chart
fig, ax = plt.subplots(figsize=(10, 6))

# Dictionary to track countries already plotted (for unique legend entries)
plotted_countries = {}

# Plot each country's emissions for 2000 and 2020
for _, row in df_pivot.iterrows():
    line, = ax.plot([2000, 2020], [row["2000"], row["2020"]], marker='o', label=row["Area"])
    plotted_countries[row["Area"]] = line  # Store only unique country labels

# Set axis labels and title
plt.xticks([2000, 2020], ['2000', '2020'])
plt.xlabel("Year")
plt.ylabel("CO₂ Emissions (Million Metric Tons)")
plt.title("Change in CO₂ Emissions by Country (2000-2020)")

# Manually add legend to ensure uniqueness
ax.legend(plotted_countries.values(), plotted_countries.keys(), loc='best', fontsize=8)

plt.grid(True, linestyle='--', alpha=0.7)
plt.show()
```
