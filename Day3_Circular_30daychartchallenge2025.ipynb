```python
# Aggregate data by Financial Class → Doctor Type → Patient Type
df_grouped = df.groupby(["Financial Class", "Doctor Type", "Patient Type"]).size().reset_index(name="count")

# Create Sunburst Chart
fig = px.sunburst(
    df_grouped,
    path=["Financial Class", "Doctor Type", "Patient Type"],  # Hierarchical levels
    values="count",
    title="Hospital Financial & Patient Classification",
    color="count",
    color_continuous_scale="Blues"
)

# Show Chart
fig.show()


```

```python


# Aggregate data by Financial Class → Doctor Type → Patient Type
df_grouped = df.groupby(["Financial Class", "Doctor Type", "Patient Type"]).size().reset_index(name="count")

# Create Sunburst Chart
fig = px.sunburst(
    df_grouped,
    path=["Financial Class", "Doctor Type", "Patient Type"],  # Hierarchical levels
    values="count",
    title="Hospital Financial & Patient Classification",
    color="count",
    color_continuous_scale="Blues"
)

# Adjust title and legend position
fig.update_layout(
    title_x=0.5,  # Center title
    margin=dict(t=50, l=0, r=0, b=0),  # Reduce whitespace
    legend=dict(
        x=0.9, y=0.9,  # Move legend closer to chart
        xanchor="right", yanchor="top",
        bgcolor="rgba(255,255,255,0.6)"  # Semi-transparent background for better readability
    )
)

# Show Chart
fig.show()
```

```python
# Aggregate data by Financial Class → Doctor Type → Patient Type
df_grouped = df.groupby(["Financial Class", "Doctor Type", "Patient Type"]).size().reset_index(name="count")

# Create Sunburst Chart
fig = px.sunburst(
    df_grouped,
    path=["Financial Class", "Doctor Type", "Patient Type"],  # Hierarchical levels
    values="count",
    title="Hospital Financial & Patient Classification",
    color="count",
    color_continuous_scale="Blues"
)

# Create Sunburst Chart
fig = px.sunburst(
    df_grouped,
    path=["Financial Class", "Doctor Type", "Patient Type"],  # Hierarchical levels
    values="count",
    title="Hospital Financial & Patient Classification",
    color="count",
    color_continuous_scale="Blues"
)

# Adjust title and move color bar closer
fig.update_layout(
    title_x=0.5,  # Center title
    margin=dict(t=50, l=0, r=0, b=0),  # Reduce whitespace
    coloraxis_colorbar=dict(
        x=0.75,  # Moves color bar closer to the chart
        y=0.5,  # Centers color bar vertically
        len=0.6,  # Adjusts color bar size
        thickness=15,  # Adjusts width of color bar
        title="Count"  # Labels the color bar
    )
)

# Show Chart
fig.show()
```
