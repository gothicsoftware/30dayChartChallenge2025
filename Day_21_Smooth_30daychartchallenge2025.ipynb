```python
import pandas as pd
import matplotlib.pyplot as plt


data = {
    'Florida City': ['St. Augustine Beach', 'Miami', 'St. Pete Beach', 'Apalachcola', 'Naples','Clearwater Beach','Sanibel Island'],
    'Count': [15, 1, 3, 2, 2, 3, 1]
}

# Create a DataFrame
df = pd.DataFrame(data)

# Plotting the bar chart
plt.figure(figsize=(10, 6))
plt.bar(df['Florida City'], df['Count'], color='skyblue')
plt.title('Beaches Paul & Tricia have looked for Fossils (Shells)')
plt.xlabel('Florida City')
plt.ylabel('# of times visited')
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import random

data = {
    'Florida City': ['St. Augustine Beach', 'Miami', 'St. Pete Beach', 'Apalachcola', 'Naples','Clearwater Beach','Sanibel Island'],
    'Count': [15, 1, 3, 2, 2, 3, 1]
}

# Create a DataFrame
df = pd.DataFrame(data)

# Sort the DataFrame by 'Count' in descending order
df_sorted = df.sort_values('Count', ascending=False)

# Generate a list of random colors for the bars
num_bars = len(df_sorted['Florida City'])
colors = ['#'+''.join([random.choice('0123456789ABCDEF') for j in range(6)]) for i in range(num_bars)]


# Plotting the bar chart
plt.figure(figsize=(12, 6))
plt.bar(df_sorted['Florida City'], df_sorted['Count'], color=colors)
plt.title('Beaches Paul & Tricia have looked for Fossils (Shells)')
plt.xlabel('Florida City')
plt.ylabel('# of times visited')
plt.xticks(rotation=45, ha='right') # Rotate x-axis labels for better readability
plt.tight_layout()
plt.show()
```
