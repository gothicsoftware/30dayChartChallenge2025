```python
import matplotlib.pyplot as plt
import numpy as np

# Average monthly temperatures for Michigan in 2024 (C)
months = [
    "Jan", "Feb", "Mar", "Apr", "May", "Jun",
    "Jul", "Aug", "Sep", "Oct", "Nov", "Dec"
]
avg_temps = [-4.2, -3.3, 1.6, 8.7, 15.5, 20.9, 23.4, 22.4, 18.8, 11.8, 5.2, -0.9]

# Convert months to numerical indices for plotting
x = np.arange(len(months))

# Create the plot
plt.figure(figsize=(8, 5))
plt.plot(x, avg_temps, marker='o', linestyle='-', color='black', label='Avg Temp (C)')

# Fill areas above and below zero
plt.fill_between(x, avg_temps, 0, where=(np.array(avg_temps) >= 0), interpolate=True, color='skyblue', alpha=0.5, label='Above 0C')
plt.fill_between(x, avg_temps, 0, where=(np.array(avg_temps) < 0), interpolate=True, color='salmon', alpha=0.5, label='Below 0C')

# Customize the plot
plt.axhline(0, color='gray', linewidth=1, linestyle='--')
plt.xticks(x, months)
plt.xlabel('Month')
plt.ylabel('Temperature (C)')
plt.title('Average Monthly Temperatures in Michigan (2024)')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
