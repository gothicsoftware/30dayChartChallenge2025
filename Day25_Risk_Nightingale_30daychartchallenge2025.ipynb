```python
# Filter for fatal crashes only
fatal_df = df[df['FATALS'] > 0]

# Group by hour and count
fatal_by_hour = (
    fatal_df.groupby('ARR_HOUR')
    .size()
    .reindex(range(24), fill_value=0)  # Ensure all 24 hours are represented
    .reset_index(name='Fatal Crashes')
)

# Plot
plt.figure(figsize=(10, 5))
sns.set_style("whitegrid")
sns.lineplot(data=fatal_by_hour, x='ARR_HOUR', y='Fatal Crashes', marker='o', linewidth=2.5)

# Customize x-axis
plt.xticks(ticks=range(24), labels=[f"{h}:00" for h in range(24)], rotation=45)
plt.xlim(0, 23)

# Labels and title
plt.title('Fatal Crashes by Hour of Day (NHTSA, 2024)', fontsize=16, pad=20)
plt.xlabel('Hour of Day', fontsize=12)
plt.ylabel('Number of Fatal Crashes', fontsize=12)

plt.tight_layout()
plt.show()
```
